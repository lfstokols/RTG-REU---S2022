```python
from pylab import figure, cm

import matplotlib.pyplot as plt
import numpy as np


def f(x1,x2):
    return x1 * np.exp(x2)

def g(x1, x2):
    return x1**2 + 2*x1*x2 - x2**2

x1_min = -2.0
x1_max = 2.0
x2_min = -2.0
x2_max = 2.0

x1, x2 = np.meshgrid(np.arange(x1_min,x1_max, 0.1), np.arange(x2_min,x2_max, 0.1))
 
y = f(x1,x2)
```

```python
def skew_grad(f, spacing):
    nbla = np.gradient(f, spacing)
    neg_del_y = nbla[0] * (-1)
    del_x = nbla[1]
    return neg_del_y, del_x
```

```python
grad = skew_grad(y, 0.1)
```

```python
def plot_field(x1, x2, grad):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    strm1 = ax.streamplot(x1, x2, grad[0], grad[1], linewidth=2, cmap='rainbow')
    fig.colorbar(strm1.lines)
    plt.show()
    print(' ')
```

```python
plot_field(x1, x2, grad)
```

```python
plot_field(x1, x2, skew_grad(g(x1, x2), 0.1))
```
