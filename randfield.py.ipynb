```python
import numpy as np
from random import random

SEED = 0
np.random.seed(SEED)

# n is the Fourier modes to cut off at
def F(x1, x2):
    n = 10
    F1 = 0
    F2 = 0
    
    for j in range(n):
        for k in range(n):
            coeffj = 2 * np.pi * j
            coeffk = 2 * np.pi * k
            
            weights = np.random.random(4)
    
            # Make weights fall between (-1, 1)
            norm = weights[0] ** 2 + weights[1] ** 2 + weights[2] ** 2 + weights[3] ** 2
            weights = [(w * 2 - 1) /norm for w in weights]
                
            F1 += - weights[0] * coeffk * np.cos(coeffj * x1) * np.cos(coeffk * x2) 
            F1 += weights[1] * coeffk * np.sin(coeffj * x1) * np.sin(coeffj * x2) 
            F1 += - weights[2] * coeffk * np.sin(coeffj * x1) * np.cos(coeffk * x2) 
            F1 += weights[3] * coeffk * np.cos(coeffj * x1) * np.sin(coeffk * x2)
            F2 += - weights[0] * coeffj * np.sin(coeffk * x1) * np.sin(coeffk * x2) 
            F2 += weights[1] * coeffj * np.cos(coeffj * x1) * np.cos(coeffk * x2) 
            F2 += - weights[2] * coeffj * np.sin(coeffj * x1) * np.cos(coeffk * x2) 
            F2 += weights[3] * coeffj * np.cos(coeffj * x1) * np.sin(coeffk * x2)
    
    return F1, F2

F1, F2= F(X, Y)
#xaxis = np.linspace(0, 4, 10)
#yaxis = np.linspace(-1, 1, 20)
#result = F(xaxis[:,None], yaxis[None,:])
#F1_comp = result[0]
#F2_comp = result[1]

x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

[Y, X] = np.meshgrid(x, y, indexing = 'ij', sparse = 'true')
F(X, Y)
```

```python
from pylab import figure, cm

import matplotlib.pyplot as plt

def plot_field(x1, x2,F1,F2):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    strm = ax.streamplot(x1, x2, F1, F2, linewidth=2, cmap='rainbow')
    fig.colorbar(strm.lines)
    plt.show()
    print(' ')

# Random seed 42
np.random.seed(42)
F1, F2 = F(X, Y)
plot_field(X, Y, F1, F2)
```
